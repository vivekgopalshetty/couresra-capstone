```python
import numpy as np
import pandas as pd
import requests
import lxml
from bs4 import BeautifulSoup
```

```python
source=requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
```

```python
soup=BeautifulSoup(source,'lxml')
print(soup.prettify())
```

```python
soup.find_all('td')
```

```python
cnt=0
dic={}
postcode=[]
borough=[]
neighbourhood=[]
```

```python
for trow in soup.find_all("td"):
    try:
        if(cnt==0):
            pst_code=trow.text
            cnt=cnt+1
            if(pst_code[0]!='M'):
                break
        elif(cnt==1):
            bor=trow.text
            cnt=cnt+1
        elif(cnt==2):
            neigh=trow.text
            cnt=0
            postcode.append(pst_code)
            borough.append(bor)
            neighbourhood.append(neigh)
    
    except:
        pass
```

```python
df=pd.DataFrame(columns=['post_code','borough','neighbourhood'])
df.head()
```

```python
li=[]
for i in (neighbourhood):
    i=i[0:len(i)-1]
    li.append(i)
```

```python
df['post_code']=postcode
df['borough']=borough
df['neighbourhood']=li
df.shape[0]
```

```python
condition=df['borough']!="Not assigned"
df=df[condition]
```

```python
df.head()
```

```python
li1=[]
for row in df.itertuples():
    li=[]
    for j in df.itertuples():
        if (row.post_code==j.post_code):
            li.append(j.neighbourhood)
    li1.append(li)  
```

```python
li2=[]
for i in li1:
    st=",".join(i)
    li2.append(st)
```

```python
print(li2)
```

```python
df=df.drop(columns=['neighbourhood'])
```
